## Función

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from sympy.abc import x

def exact_solution(count, function, h, lower_bound, upper_bound):
    f = sp.lambdify(x,function)
    n = int((upper_bound-lower_bound)/h)

    x_array = np.linspace(lower_bound,upper_bound,n+1)
    y_array = f(x_array)
    plt.plot(x_array, y_array,'bo') 
    fig_title = 'Heri, graph '+str(count)
    plt.title(fig_title)
    plt.grid()
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
    
    return
```

## Ejercicio 4a  

$$
y' = te^{3t} - 2y \\
y(0) = 0
$$

$$ y(t) = \frac{1}{5}te^{3t} - \frac{1}{25}e^{3t} + \frac{1}{25}e^{-2t}$$

```python
exact_solution('4a', '1/5*x*exp(3*x)-(1/25)*exp(3*x)+(1/25)*x*exp(-2*x)', 0.5, 0, 1)
```

## Ejercicio 4b

$$
y' = 1 + (t - y)^2 \\
y(2) = 1
$$

$$ y(t) = t + \frac{1}{1-t}$$

```python
exact_solution('4b', 'x+1/(1-x)', 0.5, 2, 3)
```

## Ejercicio 4c

$$
y' = 1 + \frac{y}{t} \\
y(1) = 2
$$

$$ y(t) = t \text{ln}(t) + 2t$$

```python
exact_solution('4c', 'x*log(x)+2*x', 0.25, 1, 2)
```

## Ejercicio 4d

$$
y' = \cos(2t) + \sin(3t) \\
y(0) = 1
$$

$$ y(t) = \frac{1}{2}\text{sin}(2t) - \frac{1}{3}\text{cos}(3t) + \frac{4}{3}$$

```python
exact_solution('4d', '(1/2)*sin(2*x)-(1/3)*cos(3*x)+(4/3)', 0.25, 0, 1)
```

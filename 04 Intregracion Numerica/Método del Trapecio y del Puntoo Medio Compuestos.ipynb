```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Método del trapecio compuesto
def trapezoidal_compound(a, b, n, f):
    h = (b - a) / n  # Tamaño del subintervalo
    x = np.linspace(a, b, n + 1)  # Puntos de división
    y = f(x)  # Evaluación de la función en los puntos
    integral = (h / 2) * (y[0] + 2 * np.sum(y[1:-1]) + y[-1])
    return integral, x, y
```

```python
# Método del punto medio compuesto
def midpoint_compound(a, b, n, f):
    h = (b - a) / n  # Tamaño del subintervalo
    x_mid = np.linspace(a + h/2, b - h/2, n)  # Puntos medios de los subintervalos
    y_mid = f(x_mid)  # Evaluación de la función en los puntos medios
    integral = h * np.sum(y_mid)
    return integral, x_mid, y_mid
```

# Ejemplo 1

```python
# Función a integrar
def f(x):
    return np.cos(x)  # Cambia esta función para probar otras

# Parámetros
a, b = 0, 1.5*np.pi  # Límites de integración
n = 6  # Número de subintervalos
```

```python
# Calcular la integral y los puntos
integral, x, y = trapezoidal_compound(a, b, n, f)

# Calcular el valor exacto de la integral
exact_integral = np.cos(b) - np.cos(a)

# Calcular el error
error = abs(exact_integral - integral)

# Mostrar la integral calculada y el error
print(f"Valor aproximado de la integral: {integral:.6f}")
print(f"Valor exacto de la integral: {exact_integral:.6f}")
print(f"Error: {error:.6f}")
```

```python

# Crear el gráfico
x_fine = np.linspace(a, b, 1000)  # Puntos finos para la curva real
y_fine = f(x_fine)

# Configurar el estilo manualmente
plt.figure(figsize=(18, 10))
plt.plot(x_fine, y_fine, color='#B32428', label='Curva $f(x) = \\ cos(x)$', linewidth=3)

# Dibujar los trapecios
for i in range(n):
    plt.fill_between([x[i], x[i+1]], [y[i], y[i+1]], color='#34493A', alpha=0.5)

# Puntos en la función
plt.plot(x, y, '*', color='#D2B081', label='Puntos de evaluación', markersize=18)

# Configurar la fuente Times New Roman
plt.rc('font', family='Times New Roman')

# Agregar detalles al gráfico
plt.title(f'\nMétodo del Trapecio Compuesto\n', fontsize=24, fontweight='bold', pad=20)
plt.xlabel('$x$', fontsize=16, fontweight='bold')
plt.ylabel('$f(x)$', fontsize=16, fontweight='bold')
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.legend(fontsize=14)
plt.grid(alpha=0.3, linestyle='--')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()

# Guardar el gráfico como PDF
plt.savefig('tcos.pdf', format='pdf')

# Mostrar el gráfico
plt.show()
```

```python
# Calcular la integral y los puntos usando el método del punto medio compuesto
integral_mid, x_mid, y_mid = midpoint_compound(a, b, n, f)

# Calcular el error
error_mid = abs(exact_integral - integral_mid)

# Mostrar la integral calculada y el error
print(f"Valor aproximado de la integral (Punto Medio): {integral_mid:.6f}")
print(f"Valor exacto de la integral: {exact_integral:.6f}")
print(f"Error (Punto Medio): {error_mid:.6f}")
```

```python
# Crear el gráfico
plt.figure(figsize=(18, 10))
plt.plot(x_fine, y_fine, color='#B32428', label='Curva $f(x) = \\ e^x$', linewidth=3)

# Dibujar los rectángulos del método del punto medio
for i in range(n):
    plt.fill_between([x[i], x[i+1]], [0, 0], [y_mid[i], y_mid[i]], color='#34493A', alpha=0.5)

# Puntos en la función
plt.plot(x_mid, y_mid, '*', color='#D2B081', label='Puntos de evaluación', markersize=18)

# Configurar la fuente Times New Roman
plt.rc('font', family='Times New Roman')

# Agregar detalles al gráfico
plt.title(f'\nMétodo del Punto Medio Compuesto\n', fontsize=24, fontweight='bold', pad=20)
plt.xlabel('$x$', fontsize=16, fontweight='bold')
plt.ylabel('$f(x)$', fontsize=16, fontweight='bold')
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.legend(fontsize=14)
plt.grid(alpha=0.3, linestyle='--')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()

# Guardar el gráfico como PDF
plt.savefig('midpointcos.pdf', format='pdf')

# Mostrar el gráfico
plt.show()
```
